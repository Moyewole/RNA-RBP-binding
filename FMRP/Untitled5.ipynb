```python
text_file = open("random1.fa", "w")
for i in mylist:
    text_file.write(i + '\n')
text_file.close()
```

```python
from collections import Counter

def most_common_substring(text, length):
    substrings = [text[i:i+length] for i in range(len(text)-length+1)]
    count = Counter(substrings)
    return count.most_common(50)

text = mystring
length = 8
result = most_common_substring(text, length)
print(f"The most common substring of length {length} is '{result[0]}' which occurs {result[1]} times.")
```
